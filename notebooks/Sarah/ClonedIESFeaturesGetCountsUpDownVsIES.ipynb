```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import os
import pandas as pd

import datetime
print(datetime.datetime.now())
```

```python
filelist = \
[\
    'D:\\LinuxShare\\Projects\\Sarah\\BWA\\DCL_SCG\\MacAndIES\\Intersect_D23_1_R1R2.trim.sort_UpDownFeatures.tsv',\
    'D:\\LinuxShare\\Projects\\Sarah\\BWA\\DCL_SCG\\MacAndIES\\Intersect_D23_2_R1R2.trim.sort_UpDownFeatures.tsv',\
    'D:\\LinuxShare\\Projects\\Sarah\\BWA\\DCL_SCG\\MacAndIES\\Intersect_D5_1_R1R2.trim.sort_UpDownFeatures.tsv',\
    'D:\\LinuxShare\\Projects\\Sarah\\BWA\\DCL_SCG\\MacAndIES\\Intersect_D5_2_R1R2.trim.sort_UpDownFeatures.tsv',\
    'D:\\LinuxShare\\Projects\\Sarah\\BWA\\DCL_SCG\\MacAndIES\\Intersect_ND7_1_R1R2.trim.sort_UpDownFeatures.tsv'\
]

depthcutoff = 100
alldiffups, alldiffdowns = [], []
fileprefixes = []
for f in filelist:
    import pygentoolbox.CompareEachIESToUpAndDownClonedFeature
    diffup, diffdown = pygentoolbox.CompareEachIESToUpAndDownClonedFeature.compare_ies_to_features_up_and_down_return_diffup_diffdown(f, depthcutoff)
    alldiffups = alldiffups + diffup
    alldiffdowns = alldiffdowns + diffdown
    path, file = os.path.split(f)
    fileprefixes = fileprefixes + [file.split('.')[0][10:]]*len(diffup)


# make strip charts
dfdiffups = pd.DataFrame(list(zip(alldiffups, fileprefixes)), columns=['DiffUpVsIES', 'FilePrefix'])
dfdiffdowns = pd.DataFrame(list(zip(alldiffups, fileprefixes)), columns=['DiffDownVsIES', 'FilePrefix'])

outpath = os.path.join(path, f + '.Up.pdf')
sns.stripplot(x='FilePrefix', y='DiffUpVsIES', data=dfdiffups, jitter=True, size=3)
plt.savefig(outpath)
plt.show()
plt.close()

outpath = os.path.join(path, f + '.Down.pdf')
sns.stripplot(x='FilePrefix', y='DiffDownVsIES', data=dfdiffdowns, jitter=True, size=3)
plt.savefig(outpath)
plt.show()
plt.close()
print(datetime.datetime.now())
```
